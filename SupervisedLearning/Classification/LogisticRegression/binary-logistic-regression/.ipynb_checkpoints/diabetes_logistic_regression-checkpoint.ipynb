This program analyzes https://www.kaggle.com/datasets/akshaydattatraykhare/diabetes-dataset/data from kaggle using Logistic Regression to predict whether a person has diabetes based on various health metrics.

This program will build a model to predict the outcome (diabetes or not) based on the input features.

About this file
===========================

Information about dataset attributes -

Pregnancies: To express the Number of pregnancies

Glucose: To express the Glucose level in blood

BloodPressure: To express the Blood pressure measurement

SkinThickness: To express the thickness of the skin

Insulin: To express the Insulin level in blood

BMI: To express the Body mass index

DiabetesPedigreeFunction: To express the Diabetes percentage

Age: To express the age

Outcome: To express the final result 1 is Yes and 0 is No




```python

import pandas as pd
import numpy as np
```

```python
Read the file to understand general structure of the file
```

```python
df = pd.read_csv('../../../../dataset/diabetes/diabetes.csv')
print("First 5 rows of the dataset:")
print(df.head())
```

```python

print(df.info())
print("\nStatistical summary of the dataset:")
# Display all columns in the summary instead of ...
pd.set_option('display.max_columns', None)
print(df.describe())
```

```python

# From the data description, we can see that some features have a minimum value of 0 which is not possible for those health metrics.
# We will replace those 0 values with mean and then handle them appropriately.
features_with_zero_as_invalid = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for feature in features_with_zero_as_invalid:
    mean_value = df[feature].replace(0, np.nan).mean()
    df[feature] = df[feature].replace(0, mean_value)

print(df.describe())
```
