#### **3.2 Handling Invalid Zero Values**
From the data description, we can see that following features have a minimum value of 0 which is not possible for those health metrics.

 'Glucose',

 'BloodPressure',

 'SkinThickness',

 'Insulin',

 'BMI'

We will replace those 0 values with mean if there are no outliers, and then handle them appropriately if the 0 values are significant.

Check for outliers in these features before replacing 0 values with mean.

````python
# Boxplot to check for outliers
import matplotlib.pyplot as plt
import seaborn as sns
features_with_zero_as_invalid = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_with_zero_as_invalid, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[feature])
    plt.title(f'Boxplot of {feature}')
    plt.tight_layout()
    plt.show()
````

````python
features_with_zero_as_invalid = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for feature in features_with_zero_as_invalid:
    zero_count = (df[feature] == 0).sum()
    print(f'% of zero values in {feature}:  {zero_count * 100 / len(df):.2f}%')


for feature in features_with_zero_as_invalid:
    mean_value = df[feature].replace(0, np.nan).mean()
    df[feature] = df[feature].replace(0, mean_value)

print(df.describe())
````
